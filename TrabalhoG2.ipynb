```python
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib inline

def exponencial(nsamples,rate):
    x = np.zeros(nsamples)
    u = np.random.sample(nsamples)
    for i in range(nsamples):
        x[i] = - math.log(1.0-u[i])/rate
    return x

def poissonHomogeneo(rate,T):
    N = 0
    tE = []
    t = 0.0
    while (t < T):
        Z = exponencial(1,rate)[0]
        if (t + Z < T):
            t = t + Z
            N += 1
            tE.append(t)
        else:
            break
    return tE


def newPolicyHolder(rateV, rate):
    ret = rateV / rate
    return ret


def lostPolicyHolder(n, rateMu, rate):
    ret = (n * rateMu) / rate
    #print(str(ret) + " lph")
    return ret
    
def claims(n, rateLambda, rate):
    ret = (n * rateLambda) / rate
    return ret


# Even List function, defines the t time of the next event 
def EL(t, rateV, n, rateMu, rateLambda, a):
    rate = (rateV + (n * rateMu) + (n * rateLambda))
    random = np.random.sample(1)[0] 
    J = 0
    nph = newPolicyHolder(rateV, rate)
    lph = lostPolicyHolder(n, rateMu, rate)
    cl = claims(n, rateLambda, rate)    
    
    if random >= 0 and random < nph:
        J = 1
    elif random > nph and random < lph:
        J = 2
    elif random > lph and random < cl:
        J = 3
    elif random > cl and random <= 1:
        J = 0
    return J

def test(a0, n0, pay,ratePrice, T, rateV, rateMu, rateLambda):
    t = 0.0
    a = a0
    n = n0
    c = pay
    rate = (rateV + (n * rateMu) + (n * rateLambda))
    X = exponencial(1, rate)[0]
    te = X

    while(1):
        if te > T:
            I = 1
            break
        else:
            a = a + n * c * (te - t)
            t = te

            J = EL(t, rateV, n, rateMu, rateLambda, a)        
            #print(str(J) + " aaa")
            if J == 1:
                n += 1
            if J == 2:
                n -= 1
               
            if J == 3:
                Y = exponencial(1,ratePrice)[0]
                #print(Y)
                if Y > a:
                    I = 0
                    break
                else:      
                    a = a - Y
            
            rate = (rateV + (n * rateMu) + (n * rateLambda))
            X = exponencial(1, rate)[0]    
            te = t + X
    
    return I

# Number of Claims
# Poisson
rateLambda = 10
#Rate of the exponencial distribution that generate Price of claims
ratePrice = 1/2000

# Number of New customers
# Poisson
rateV = 8

# Time the customer stays with the company 
#  Exponentially distributed
rateMu = 6
# insurance payment
pay = 11000
# Variable total time of simulation
T = 365

# System variable
# (n, a) => policyholders, firm's capital

#Capital Inicial
a = 25000
#Clientela Inicial
n = 10000


simulacoes = 1000

soma = 0
for i in range(simulacoes):
    parcial = test(a,n,pay,ratePrice,T,rateV,rateMu,rateLambda)
    soma += parcial

prob = (1 - (soma/simulacoes)) * 100
print("Probabilidade da empresa entrar em ruína no período de " + str(T) + " dias: "+ str(prob)+ "%")
```
